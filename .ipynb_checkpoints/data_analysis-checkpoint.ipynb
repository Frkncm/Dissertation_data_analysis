```python
# Read the csv file
GAME_TIME_FINISH = '-1 day'
SEC_CONVERSION = 0.78

for fn in range(1,9):   
    data_with = pd.read_csv('processed_participants_dataset/subject_' + str(fn) + '_with_shared_control_processed.csv')
    data_without = pd.read_csv('processed_participants_dataset/subject_' + str(fn) + '_without_shared_control_processed.csv')

    #start from the experiment starting time
    #data processing
    baseline_hr_with = data_with['mean_hr'][120]
    baseline_scl_with = data_with['mean_scl'][120]
    baseline_hr_without = data_without['mean_hr'][120]
    baseline_scl_without = data_without['mean_scl'][120]
    data_with = data_with[121:]
    data_without = data_without[121:]


    #data with shared control
    #find the fisrt index
    print('\n***************************************************************************\n')
    print('data with shared control subject_' + str(fn))
    #data with shared control
    #find the fisrt index
    game_time = data_with["game_time"].to_numpy()
    start_index = 0
    for i, val in enumerate(game_time):
        if val != '0:05:00':
            start_index = i - 1 
            break

    #find last index
    game_time = data_with["game_time"][start_index:].to_numpy()
    score = data_with["current_score"][start_index:].to_numpy()
    last_index = 0
    for i in range(len(game_time)):
        if (game_time[i].find(GAME_TIME_FINISH) != -1): 
            last_index = i + start_index
            print("time is up!")
            break
        if score[i] == '0' or score[i] == '0.0':
            last_index = i + start_index
            print("score is 0!")
            break

    if last_index == 0:
        last_index = i
        print("Succesfuly finished!")        

    print(experiment_start_indx)
    print(last_index)
    print()

    score_with = data_with['current_score'].to_numpy()[-1]
    last_exp_time_with = data_with['game_time'].to_numpy()[-1]
    if(last_exp_time_with.find(GAME_TIME_FINISH) != -1): 
        last_exp_time_with = 0
    game_time_with = data_with["game_time"][start_index:last_index].to_numpy()
    gsr_scl_values_with = data_with["mean_scl"][start_index:last_index].to_numpy()
    gsr_scr_values_with = data_with["mean_scr"][start_index:last_index].to_numpy()
    hr_values_with = data_with["current_hr"][start_index:last_index].to_numpy() 
    stress_time_with= data_with["is_stressed"][start_index:last_index].to_numpy() 

    #data without shared control
    #find the fisrt index
    print('data without shared control')
    game_time = data_without["game_time"].to_numpy()
    start_index = 0
    for i, val in enumerate(game_time):
        if val != '0:05:00':
            start_index = i - 1
            break

    #find last index
    game_time = data_without["game_time"][start_index:].to_numpy()
    score = data_without["current_score"][start_index:].to_numpy()
    last_index = 0
    for i in range(len(game_time)):
        if (game_time[i].find(GAME_TIME_FINISH) != -1): 
            last_index = i + start_index
            print("time is up!")
            break
        if score[i] == '0' or score[i] == '0.0':
            last_index = i + start_index
            print("score is 0!")
            break


    if last_index == 0:
        last_index = i
        print("Succesfuly finished!")

    print(start_index)
    print(last_index)
    print()

    
    score_without = data_without['current_score'].to_numpy()[-1]
    last_exp_time_without = data_without['game_time'].to_numpy()[-1]
    if(last_exp_time_without.find(GAME_TIME_FINISH) != -1): 
        last_exp_time_without = 0
    game_time_without = data_without["game_time"][start_index:last_index].to_numpy()
    gsr_scl_values_without = data_without["mean_scl"][start_index:last_index].to_numpy()
    gsr_scr_values_without = data_without["mean_scr"][start_index:last_index].to_numpy()
    hr_values_without = data_without["current_hr"][start_index:last_index].to_numpy() 
    stress_time_without = data_without["is_stressed"][start_index:last_index].to_numpy() 

    
    print('\nwith shared control')
    print('baseline hr: {}'.format(baseline_hr_with))
    print('baseline scl: {}'.format(baseline_scl_with))
    print('mean hr:{}  std hr: {}' .format(np.mean(hr_values_with.astype('float32')), 
                                         np.std(hr_values_with.astype('float32'), axis=0)))
    print('mean scl:{}  std scl: {}' .format(np.mean(gsr_scl_values_with.astype('float32')), 
                                         np.std(gsr_scl_values_with.astype('float32'), axis=0)))
    print('mean scr:{}  std scr: {}' .format(np.mean(gsr_scr_values_with.astype('float32')), 
                                         np.std(gsr_scr_values_with.astype('float32'), axis=0)))
    print('score: {}'.format(score_with))
    print('last game time {}'.format(last_exp_time_with))
    print('stressed time: {}'.format(len(np.where(stress_time_with == '1')[0])*SEC_CONVERSION))
    print('not-stressed time: {}'.format(len(np.where(stress_time_with == '0')[0])*SEC_CONVERSION))
    
    
    
    print('\nwithout shared control')
    print('baseline hr: {}'.format(baseline_hr_without))
    print('baseline scl: {}'.format(baseline_scl_without))
    print('mean hr:{}  std hr: {}' .format(np.mean(hr_values_without.astype('float32')), 
                                         np.std(hr_values_without.astype('float32'), axis=0)))
    print('mean scl:{}  std scl: {}' .format(np.mean(gsr_scl_values_without.astype('float32')), 
                                         np.std(gsr_scl_values_without.astype('float32'), axis=0)))
    print('mean scr:{}  std scr: {}' .format(np.mean(gsr_scr_values_without.astype('float32')), 
                                         np.std(gsr_scr_values_without.astype('float32'), axis=0)))
    print('score: {}'.format(score_without))
    print('last game time {}'.format(last_exp_time_without))
    print('stressed time: {}'.format(len(np.where(stress_time_without == '1')[0])*SEC_CONVERSION))
    print('not-stressed time: {}'.format(len(np.where(stress_time_without == '0')[0])*SEC_CONVERSION))
    
    
    #distribution graph
    sns.distplot(hr_values_with, label='with shared control')
    sns.distplot(hr_values_without, label='without shared control')
    plt.legend()
    plt.show()


    #distribution graph
    sns.distplot(gsr_scl_values_with, label='with shared control')
    sns.distplot(gsr_scl_values_without, label='without shared control')
    plt.legend()
    plt.show()

    #distribution graph
    sns.distplot(gsr_scr_values_with, label='with shared control')
    sns.distplot(gsr_scr_values_without, label='without shared control')
    plt.legend()
    plt.show()
```

```python
#plot distribution graphs of "interactions" for with_robot and without robot conditions to decide which test we need to choose
sns.distplot(df[df.exp_type == "with_shared_control"]["mean_hr"], label='with_shared_control')
sns.distplot(df[df.exp_type == "without_shared_control"]["mean_hr"], label='without_shared_control')
plt.legend()
plt.show()

#As can be seen in the distribution graph, we can use t-test. So, show the t-test results
print("mean_hr:")
print(ttest_ind(df[df.exp_type == "with_shared_control"].mean_hr, df[df.exp_type == "without_shared_control"].mean_hr))

# non-parametric u-test
print("\nMann-Whitney U Test")
stat, p = mannwhitneyu(df[df.exp_type == "with_shared_control"].mean_hr, df[df.exp_type == "without_shared_control"].mean_hr)
print('mean_hr\nStatistics=%.3f, p=%.3f' % (stat, p))

# non-parametric u-test
print("\nWilcoxon Signed-Rank Test")
stat, p = wilcoxon(df[df.exp_type == "with_shared_control"].mean_hr, df[df.exp_type == "without_shared_control"].mean_hr)
print('mean_hr\nStatistics=%.3f, p=%.3f' % (stat, p))
```
